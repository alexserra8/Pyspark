# 2. Creating SparkSession

One aspect of the explanation why SparkSession is preferable over SparkContext in SparkSession Vs SparkContext battle is that SparkSession unifies all of Spark’s numerous contexts, removing the developer’s need to worry about generating separate contexts.

```python
#Create the SparkSession
my_spark = SparkSession.builder.getOrCreate()

#print the session
print(my_spark)
```

# 3. Performing Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= my_spark.read.csv("insurance.csv", inferSchema=True, header=True)
```

```python
df.show(5)
```

```python
df.describe().show()
```

```python
#Transforming to pandas dataframe to plot with matplolib and seaborn
pandas_df = df.toPandas()
```

```python
#Let's make some plot to have some idea
for e in pandas_df.columns:
    sns.histplot(data=pandas_df, x=e)
    plt.title(e)
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
sns.histplot(data=pandas_df, x="age", ax=axes[0,0])
axes[0,0].set_title("Age" + " vs Expenses")

sns.histplot(data=pandas_df, x="sex", ax=axes[0,1], color="tab:green")
axes[0,1].set_title("Sex" + " vs Expenses")

sns.histplot(data=pandas_df, x="bmi", ax=axes[1,0], color="tab:red")
axes[1,0].set_title("bmi" + " vs Expenses")

sns.histplot(data=pandas_df, x="children", ax=axes[1,1], color="tab:orange")
axes[1,1].set_title("Children" + " vs Expenses")

sns.histplot(data=pandas_df, x="smoker", ax=axes[2,0], color="tab:olive")
axes[2,0].set_title("Smoker" + " vs Expenses")

sns.histplot(data=pandas_df, x="region", ax=axes[2,1], color="tab:purple")
axes[2,1].set_title("Region" + " vs Expenses")

```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.boxplot(data=pandas_df, x="expenses", y="sex", dodge=False, orient="h", ax=axes[0,0])
axes[0,0].set_title("Sex" + " vs Expenses")

sns.boxplot(data=pandas_df, x="expenses", y="smoker", dodge=False, orient="h", ax=axes[0,1])
axes[0,1].set_title("Smoker" + " vs Expenses")

sns.boxplot(data=pandas_df, x="expenses", y="region", dodge=False, orient="h", ax=axes[1,0])
axes[1,0].set_title("Region" + " vs Expenses")
```

# 5. Using StringIndexer for dealing with Categorical data

```python
from pyspark.ml.feature import StringIndexer
```

```python
indexer = StringIndexer(inputCol="sex", outputCol="gender_cat")
indexed=indexer.fit(df).transform(df)
indexed.show(3)
```

```python
indexer = StringIndexer(inputCol="smoker", outputCol="smoker_cat")
indexed=indexer.fit(indexed).transform(indexed)
indexed.show(3)
```

```python
indexer = StringIndexer(inputCol="region", outputCol="region_cat")
indexed=indexer.fit(indexed).transform(indexed)
indexed.show(3)
```
